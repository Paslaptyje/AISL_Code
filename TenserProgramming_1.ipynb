<h3>2. list_A에 50 미만의(0포함) 양의 정수 중 홀수만 넣어서 출력하시오</h3>

```python
list_A = []

for i in range (0,51):    # 2로 나누어 떨어지면 버리기
    if i % 2 != 0:
        list_A.append(i)

print(list_A)
```

<h3>3. cart는 비어있는 상태입니다. cart에 '양배추','우유','닭가슴살','제로콜라','단백질바'를 추가하고 계산대에 올리시오(출력하시오).
<br>그러나 예산초과로 '단백질바'를 제외하고 계산대에 올리시오(출력하시오). → if문을 사용하세요<br>
계산대에 올린 cart 내용물을 정렬 후 출력하시오</h3>

```python
cart_list = ['양배추','우유','닭가슴살','제로콜라','단백질바']
cart = []
len_list = len(cart_list)
print(cart_list)

for i in range(0, len_list):    # 새로운 리스트에 품목 저장
    cart.append(cart_list[i])
    if cart[i] == '단백질바':   # 저장할 때 단백질바 제거
        cart.pop()
print(cart)

cart.sort()   # 정렬
print(cart)
```

<h3>5. 2차원 리스트 동적 생성하시오 → 1차원 리스트끼리 합치게 1개, 2차원 리스트끼리 합치게 1개</h3>

```python
row = 3
column = 5
a = []

for row in range(row):
    a += [0]*column

print(a)

row = 3
column = 5
b = []

for row in range(row):
    b += [[0]*column]

print(b)
```

<h3>6. 알파벳 대소문자로 된 단어가 주어지면, 이 단어에서 가장 많이 사용된 알파벳이 무엇인지 알아내는 프로그램을 작성하시오. 단, 대문자로 출력되게 하시오</h3>

```python
import string
word = input().lower()
word_dic = {}


alphabet_list = list(string.ascii_lowercase)
# 알파벳 리스트 생성
# alphabet_list = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

len_alphabet_list = len(alphabet_list)

for char in alphabet_list:      # {key = 알파벳, value = 0} 으로 딕셔너리 생성
    word_dic[char] = 0


for i in word:                  # word에 있는 알파벳과 word_dic 있는 딕셔너리 원소가 같으면 카운트 수 증가
    if i in word_dic:
        word_dic[i] += 1


max_key = max(word_dic, key=word_dic.get)       # max()를 이용하여 최대값을 가지는 key값 반환
print(max_key.upper())

```

<h3>7. "a,b,c,d,e,f"를 ['a', 'b', 'c,d,e,f']으로 2가지 방법으로 출력하시오</h3>

```python
text = map(str, "a,b,c,d,e,f".split(','))
text_temp = list(text)        # 텍스트를 알파벳으로 나눠 리스트를 생성
text_1 = []

text_1.append(text_temp[0])   # text_1 에 알파벳 a, b를 저장
text_1.append(text_temp[1])

last_text = []

for i in range(2,6):
        last_text.append(text_temp[i])    # last_text에 나머지 알파벳 저장

temp = ",".join(last_text)    # ,를 구분 기준으로 리스트 원소 병합
text_1.append(temp)      # text_1 에 temp 추가

print(text_1)

"""

"""

input_text = map(str, "a,b,c,d,e,f".split(','))
text_2 = list(input_text)

first_list = text_2[0:1]        # a 슬라이싱

second_list = text_2[1:2]       # b 슬라이싱

third_list = text_2[2:6]        # c d e f 슬라이싱
third_list_temp = ",".join(third_list)      # ,를 구분 기준으로 리스트 원소 병합

sum_list = first_list + second_list
sum_list.append(third_list_temp)

print(sum_list)
```
